```python
import pandas as pd
import numpy as np


from sklearn import preprocessing
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('World-happiness-report-2024.csv')
df
```

```python
df.columns = df.columns.str.lower()
df
```

```python
df.isna().sum()
df= df.dropna()
```

```python
df.isna().sum()
```

```python
df = pd.get_dummies(df, columns=['country name', 'regional indicator'])
```

```python
features = df.drop('ladder score', axis=1)
target = df['ladder score']
```

```python
features_train, features_valid, target_train, train_valid = train_test_split(features, target, test_size=0.2, random_state=42)
features_test, features_valid, target_valid, target_test = train_test_split(features, target,test_size=0.5, random_state=42)
```

```python
display(features_train.shape)
display(features_valid.shape)
display(features_test.shape)
display(target_train.shape)
display(target_valid.shape)
display(target_test.shape)
```

```python
print(df['ladder score'].dtype)
print(df['ladder score'].nunique())
```

```python

best_model = None
best_result = 10
best_est = 0
best_depth = 0
for est in range(1, 10, 1):
    for depth in range (1, 5):
        model_rtf = RandomForestRegressor(random_state=12345, n_estimators=est, max_depth=depth)
        model_rtf.fit(features_train, target_train)
        pred_valid = model_rtf.predict(features_valid)
        result = mean_squared_error(target_valid, pred_valid)**0.5
        if result < best_result:
            best_model = model_rtf
            best_result = result
            best_est = est
            best_depth = depth
print('RMSE of the best model on the validation set:', best_result, 'n_stimators:', best_est, 'best_depth', depth)
```

```python
from sklearn.metrics import r2_score

model_rtf = RandomForestRegressor(random_state=12345, n_estimators=2, max_depth=4)
model_rtf.fit(features_train, target_train)
pred_valid = model_rtf.predict(features_valid)
model_rtf.score(features,target)


```
